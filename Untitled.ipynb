```python
import pandas as pd
```

```python
volunteersList = {}
columnsList = [
    "Record Time (Records each 0.5s)",
    "Date (Experiment Date)",
    "Signal level (0 (perfect) -200 (no signal))",
    "Attention (0-100% (maximum attention))",
    "Meditation (0-100% (maximum meditation))",
    "Delta (1-3Hz)",
    "Theta (4-7Hz)",
    "Low Alpha (8-9Hz)",
    "High Alfa (10-12Hz)",
    "Low Beta (13-17Hz)",
    "High Beta (18-30Hz)",
    "Low Gama (31-40Hz)",
    "High Gama (41-50Hz)",
    "RAW EEG (Voltage)",
    "Blink Strength (0-255 (maximum))",
    "Counter (not used)",
    "Average Attention (0-100% (maximum attention))",
    "Counter (not used)",
]
```

```python
count = 1
while (count < 11):
    key = 'volunteer' + str(count)
    value = pd.read_csv("./ExperimentData/volunteer01.txt", sep=";")
    value.columns = columnsList
    value.drop("Counter (not used)", axis=1)
    volunteersList[key] = value

    count += 1
```
